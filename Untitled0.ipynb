```python
# Cell 2: loading data
def load_ultrasound_data(data_path, image_size=(128, 128)):
    """Load ultrasound .tif images and exclude mask files based on naming convention"""

    # Get all .tif files in the directory
    all_files = [f for f in os.listdir(data_path) if f.lower().endswith('.tif')]

    # Filter out mask files (files containing '_mask' in the filename)
    scan_files = [f for f in all_files if '_mask' not in f.lower()]
    mask_files = [f for f in all_files if '_mask' in f.lower()]

    # Sort for consistent ordering
    scan_files = sorted(scan_files)

    print(f"Directory: {data_path}")
    print(f"Found {len(all_files)} total .tif files")
    print(f"  - Scan files: {len(scan_files)} (will be loaded)")
    print(f"  - Mask files: {len(mask_files)} (will be excluded)")

    # Show sample filenames for verification
    if scan_files:
        print(f"Sample scan files: {scan_files[:5]}")
    if mask_files:
        print(f"Sample mask files: {mask_files[:5]}")

    images = []
    skipped_files = []

    print(f"Loading {len(scan_files)} scan images...")

    for img_name in tqdm(scan_files, desc="Loading images"):
        try:
            img_path = os.path.join(data_path, img_name)
            img = Image.open(img_path).convert('L')  # Convert to grayscale
            img = img.resize(image_size, Image.BILINEAR)
            arr = np.asarray(img, dtype=np.float32) / 255.0  # Normalize to [0,1]
            images.append(arr)
        except Exception as e:
            print(f"Error loading {img_name}: {e}")
            skipped_files.append(img_name)
            continue

    if skipped_files:
        print(f"âš ï¸  Skipped {len(skipped_files)} files due to errors: {skipped_files}")

    print(f"âœ… Successfully loaded {len(images)} ultrasound images")
    print(f"Image shape: {images[0].shape if images else 'No images loaded'}")

    return np.array(images)

def analyze_filename_pattern(data_path):
    """Analyze the filename patterns in the directory"""
    files = [f for f in os.listdir(data_path) if f.lower().endswith('.tif')]

    print(f"Analyzing filename patterns in {data_path}:")
    print(f"Total .tif files: {len(files)}")

    # Separate scans and masks
    scans = [f for f in files if '_mask' not in f.lower()]
    masks = [f for f in files if '_mask' in f.lower()]

    print(f"Scan files: {len(scans)}")
    print(f"Mask files: {len(masks)}")

    # Show patterns
    if scans:
        print(f"Sample scan filenames:")
        for i, scan in enumerate(scans[:10]):  # Show first 10
            print(f"  {i+1}. {scan}")

    if masks:
        print(f"Sample mask filenames:")
        for i, mask in enumerate(masks[:10]):  # Show first 10
            print(f"  {i+1}. {mask}")

    return scans, masks
```

```python
# Cell 4: Rayleigh Noise Function (Simplified)
# ---------------------------------------------
def add_rayleigh_noise(img, scale=0.2):
    """Add Rayleigh noise to a single image"""
    noise_rayleigh = np.random.rayleigh(scale=scale, size=img.shape)
    noisy = img * (1 + noise_rayleigh)
    noisy = np.clip(noisy, 0.0, 1.0)
    return noisy

# Test the noise function
def preview_rayleigh_noise(test_img, scales=[0.05, 0.25, 0.5, 0.75]):
    """Preview how Rayleigh noise looks at different scales"""
    fig, axes = plt.subplots(1, len(scales) + 1, figsize=(15, 4))

    axes[0].imshow(test_img, cmap='gray')
    axes[0].set_title('Clean Image')
    axes[0].axis('off')

    for i, scale in enumerate(scales):
        noisy = add_rayleigh_noise(test_img, scale=scale)
        axes[i+1].imshow(noisy, cmap='gray')
        axes[i+1].set_title(f'Rayleigh Ïƒ={scale}')
        axes[i+1].axis('off')

    plt.tight_layout()
    plt.show()
```

```python
# Cell 5
class RayleighNoiseDataset(Dataset):
    def __init__(self, clean_images, noise_scales):
        self.clean_images = clean_images
        self.noise_scales = noise_scales
        # Total length = num_images * num_noise_scales
        self.total_length = len(clean_images) * len(noise_scales)

    def __len__(self):
        return self.total_length

    def __getitem__(self, idx):
        # Determine which image and which noise scale
        img_idx = idx // len(self.noise_scales)
        noise_idx = idx % len(self.noise_scales)

        # Get clean image
        clean_img = self.clean_images[img_idx]

        # Generate Rayleigh noisy image on-the-fly
        scale = self.noise_scales[noise_idx]
        noisy_img = add_rayleigh_noise(clean_img, scale=scale)

        # Convert to tensors
        clean_tensor = torch.FloatTensor(clean_img).unsqueeze(0)  # (1, H, W)
        noisy_tensor = torch.FloatTensor(noisy_img).unsqueeze(0)  # (1, H, W)

        return noisy_tensor, clean_tensor
```

```python
# Cell 6: Model Architectures (Updated with CNNAutoencoderLarge)
# ----------------------------------------------------------------
class UltrasoundAutoencoder(nn.Module):
    def __init__(self):
        super(UltrasoundAutoencoder, self).__init__()
        # Encoder
        self.encoder = nn.Sequential(
            nn.Conv2d(1, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, 2),
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, 2),
        )
        # Decoder
        self.decoder = nn.Sequential(
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Upsample(scale_factor=2, mode='nearest'),
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Upsample(scale_factor=2, mode='nearest'),
            nn.Conv2d(64, 1, kernel_size=3, padding=1),
            nn.Sigmoid()
        )

    def forward(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded

class CNNAutoencoderLarge(nn.Module):
    def __init__(self):
        super(CNNAutoencoderLarge, self).__init__()
        # Encoder
        self.encoder = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=3, padding=1),   # (B, 32, 128, 128)
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, 2),                           # (B, 32, 64, 64)
            nn.Conv2d(32, 32, kernel_size=3, padding=1),  # (B, 32, 64, 64)
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, 2),                           # (B, 32, 32, 32)
            nn.Conv2d(32, 32, kernel_size=3, padding=1),  # (B, 32, 32, 32)
            nn.ReLU(inplace=True),
        )
        # Decoder
        self.decoder = nn.Sequential(
            nn.Upsample(scale_factor=2, mode='nearest'),  # (B, 32, 64, 64)
            nn.Conv2d(32, 32, kernel_size=3, padding=1),  # (B, 32, 64, 64)
            nn.ReLU(inplace=True),
            nn.Upsample(scale_factor=2, mode='nearest'),  # (B, 32, 128, 128)
            nn.Conv2d(32, 1, kernel_size=3, padding=1),   # (B, 1, 128, 128)
            # Linear activation (no sigmoid)
        )

    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x
```

```python
# Cell 7: Simplified Training Function (No Mixed Precision Issues)
# ---------------------------------------------------------------
def train_model_optimized(model, train_loader, criterion, optimizer, device, num_epochs=50):
    model.train()
    loss_history = []

    for epoch in range(num_epochs):
        epoch_loss = 0.0
        total_psnr = 0.0
        total_ssim = 0.0
        count = 0

        progress_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs}")

        for batch_idx, (noisy, clean) in enumerate(progress_bar):
            noisy, clean = noisy.to(device, non_blocking=True), clean.to(device, non_blocking=True)

            optimizer.zero_grad()

            # Standard forward pass (no mixed precision to avoid errors)
            outputs = model(noisy)
            loss = criterion(outputs, clean)

            # Standard backward pass
            loss.backward()
            optimizer.step()

            epoch_loss += loss.item()

            # Compute metrics less frequently for speed (every 50 batches)
            if batch_idx % 50 == 0 and batch_idx > 0:
                with torch.no_grad():
                    # Only compute for first sample in batch for maximum speed
                    output_sample = outputs[0, 0].cpu().numpy()
                    clean_sample = clean[0, 0].cpu().numpy()

                    psnr = peak_signal_noise_ratio(clean_sample, output_sample, data_range=1.0)
                    ssim = structural_similarity(clean_sample, output_sample, data_range=1.0)
                    total_psnr += psnr
                    total_ssim += ssim
                    count += 1

            # Update progress bar every 25 batches for better performance
            if batch_idx % 25 == 0:
                current_psnr = total_psnr / max(count, 1)
                current_ssim = total_ssim / max(count, 1)
                progress_bar.set_postfix({
                    'Loss': f'{loss.item():.4f}',
                    'PSNR': f'{current_psnr:.1f}',
                    'SSIM': f'{current_ssim:.3f}'
                })

        avg_loss = epoch_loss / len(train_loader)
        avg_psnr = total_psnr / max(count, 1)
        avg_ssim = total_ssim / max(count, 1)
        loss_history.append(avg_loss)

        print(f"Epoch [{epoch+1}/{num_epochs}] Loss: {avg_loss:.6f} | PSNR: {avg_psnr:.2f} dB | SSIM: {avg_ssim:.4f}")

        # Clear cache every 10 epochs
        if epoch % 10 == 0 and torch.cuda.is_available():
            torch.cuda.empty_cache()

    return loss_history
```

```python
# Cell 8: Rayleigh Noise Evaluation Function
# -------------------------------------------
def evaluate_rayleigh_denoising(model, test_images, device, scale, num_examples=3):
    """Evaluate model performance on Rayleigh noise at specific scale"""
    model.eval()
    psnr_list = []
    ssim_list = []

    idxs = random.sample(range(len(test_images)), min(num_examples, len(test_images)))

    fig, axes = plt.subplots(3, num_examples, figsize=(4 * num_examples, 8))
    if num_examples == 1:
        axes = axes.reshape(-1, 1)

    for i, idx in enumerate(idxs):
        clean = test_images[idx]
        noisy = add_rayleigh_noise(clean, scale=scale)

        noisy_tensor = torch.FloatTensor(noisy).unsqueeze(0).unsqueeze(0).to(device)

        with torch.no_grad():
            denoised_tensor = model(noisy_tensor)
            denoised = denoised_tensor.squeeze().cpu().numpy()

        psnr = peak_signal_noise_ratio(clean, denoised, data_range=1.0)
        ssim = structural_similarity(clean, denoised, data_range=1.0)
        psnr_list.append(psnr)
        ssim_list.append(ssim)

        axes[0, i].imshow(clean, cmap='gray')
        axes[0, i].set_title('Clean Image')
        axes[0, i].axis('off')

        axes[1, i].imshow(noisy, cmap='gray')
        axes[1, i].set_title(f'Rayleigh Noise (Ïƒ={scale})')
        axes[1, i].axis('off')

        axes[2, i].imshow(denoised, cmap='gray')
        axes[2, i].set_title(f'Denoised\nPSNR: {psnr:.2f} dB\nSSIM: {ssim:.3f}')
        axes[2, i].axis('off')

    plt.suptitle(f'Rayleigh Noise Denoising Results (Ïƒ={scale})')
    plt.tight_layout()
    plt.show()

    return np.mean(psnr_list), np.mean(ssim_list)
```

```python
# Cell 9b: CNN Training Configuration
# -----------------------------------
print("ðŸ”„ SWITCHING TO CNN ARCHITECTURE")
# Add this at the start of Cell 9b, before "print("ðŸ”„ SWITCHING TO CNN ARCHITECTURE")"

# Quick data loading if not already done
if 'train_dataset' not in locals():
    print("Loading data first...")
    train_images = load_ultrasound_data(train_path, image_size=(128, 128))
    test_images = load_ultrasound_data(test_path, image_size=(128, 128))

    rayleigh_scales = [0.05, 0.25, 0.5, 0.75]
    train_dataset = RayleighNoiseDataset(train_images, rayleigh_scales)
    train_loader = DataLoader(train_dataset, batch_size=96, shuffle=True, num_workers=0, pin_memory=False)
    print(f"âœ… Dataset created: {len(train_dataset)} samples")

# Use the same configuration as before
rayleigh_scales = [0.05, 0.25, 0.5, 0.75]
num_epochs = 100
batch_size = 96
learning_rate = 0.001

# Create CNN model instead of UltrasoundAutoencoder
print("Initializing CNNAutoencoderLarge...")
model_cnn = CNNAutoencoderLarge().to(device)
criterion = nn.MSELoss()
optimizer = optim.Adam(model_cnn.parameters(), lr=learning_rate)

# Learning rate scheduler
scheduler = optim.lr_scheduler.ReduceLROnPlateau(
    optimizer, mode='min', factor=0.8, patience=5, verbose=True, min_lr=1e-6
)

# Use the same dataset and dataloader
print("Using existing dataset and dataloader...")
print(f"Dataset size: {len(train_dataset)} samples")
print(f"Batch size: {batch_size}")

# Print CNN model summary
total_params = sum(p.numel() for p in model_cnn.parameters())
trainable_params = sum(p.numel() for p in model_cnn.parameters() if p.requires_grad)
print(f"\nCNN Model Summary:")
print(f"Total parameters: {total_params:,}")
print(f"Trainable parameters: {trainable_params:,}")

# Check GPU memory
if torch.cuda.is_available():
    print(f"GPU Memory: {torch.cuda.get_device_properties(0).total_memory / 1024**3:.1f} GB")
    torch.cuda.empty_cache()  # Clear any previous model from memory
    print(f"GPU Memory after cleanup: {torch.cuda.memory_allocated() / 1024**3:.2f} GB allocated")

print("\nðŸš€ CNN TRAINING CONFIGURATION:")
print(f"ðŸ“Š Architecture: CNNAutoencoderLarge")
print(f"âš¡ Parameters: {total_params:,} (fewer than UltrasoundAutoencoder)")
print(f"ðŸŽ¯ Output: Linear activation (no sigmoid)")
print(f"ðŸ”§ Channels: 1â†’32â†’32â†’1")
print(f"âœ… Ready to train CNN model!")
```

```python
# Cell 10
print("ðŸš€ Starting OPTIMIZED training...")
print("-" * 50)

loss_history = train_model_optimized(model_cnn, train_loader, criterion, optimizer, device, num_epochs=num_epochs)

# Plot loss curve
plt.figure(figsize=(10, 6))
plt.plot(loss_history)
plt.title('Optimized Training Loss Curve')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.show()

print("âœ… Optimized training completed!")
```

```python
# Cell 2:
def load_ultrasound_data(data_path, image_size=(128, 128)):
    """Load ultrasound .tif images and exclude mask files based on naming convention"""

    # Get all .tif files in the directory
    all_files = [f for f in os.listdir(data_path) if f.lower().endswith('.tif')]

    # Filter out mask files (files containing '_mask' in the filename)
    scan_files = [f for f in all_files if '_mask' not in f.lower()]
    mask_files = [f for f in all_files if '_mask' in f.lower()]

    # Sort for consistent ordering
    scan_files = sorted(scan_files)

    print(f"Directory: {data_path}")
    print(f"Found {len(all_files)} total .tif files")
    print(f"  - Scan files: {len(scan_files)} (will be loaded)")
    print(f"  - Mask files: {len(mask_files)} (will be excluded)")

    # Show sample filenames for verification
    if scan_files:
        print(f"Sample scan files: {scan_files[:5]}")
    if mask_files:
        print(f"Sample mask files: {mask_files[:5]}")

    images = []
    skipped_files = []

    print(f"Loading {len(scan_files)} scan images...")

    for img_name in tqdm(scan_files, desc="Loading images"):
        try:
            img_path = os.path.join(data_path, img_name)
            img = Image.open(img_path).convert('L')  # Convert to grayscale
            img = img.resize(image_size, Image.BILINEAR)
            arr = np.asarray(img, dtype=np.float32) / 255.0  # Normalize to [0,1]
            images.append(arr)
        except Exception as e:
            print(f"Error loading {img_name}: {e}")
            skipped_files.append(img_name)
            continue

    if skipped_files:
        print(f"âš ï¸  Skipped {len(skipped_files)} files due to errors: {skipped_files}")

    print(f"âœ… Successfully loaded {len(images)} ultrasound images")
    print(f"Image shape: {images[0].shape if images else 'No images loaded'}")

    return np.array(images)

def analyze_filename_pattern(data_path):
    """Analyze the filename patterns in the directory"""
    files = [f for f in os.listdir(data_path) if f.lower().endswith('.tif')]

    print(f"Analyzing filename patterns in {data_path}:")
    print(f"Total .tif files: {len(files)}")

    # Separate scans and masks
    scans = [f for f in files if '_mask' not in f.lower()]
    masks = [f for f in files if '_mask' in f.lower()]

    print(f"Scan files: {len(scans)}")
    print(f"Mask files: {len(masks)}")

    # Show patterns
    if scans:
        print(f"Sample scan filenames:")
        for i, scan in enumerate(scans[:10]):  # Show first 10
            print(f"  {i+1}. {scan}")

    if masks:
        print(f"Sample mask filenames:")
        for i, mask in enumerate(masks[:10]):  # Show first 10
            print(f"  {i+1}. {mask}")

    return scans, masks
```

```python
# Cell 4: Rayleigh Noise Function (Simplified)
# ---------------------------------------------
def add_rayleigh_noise(img, scale=0.2):
    """Add Rayleigh noise to a single image"""
    noise_rayleigh = np.random.rayleigh(scale=scale, size=img.shape)
    noisy = img * (1 + noise_rayleigh)
    noisy = np.clip(noisy, 0.0, 1.0)
    return noisy

# Test the noise function
def preview_rayleigh_noise(test_img, scales=[0.05, 0.25, 0.5, 0.75]):
    """Preview how Rayleigh noise looks at different scales"""
    fig, axes = plt.subplots(1, len(scales) + 1, figsize=(15, 4))

    axes[0].imshow(test_img, cmap='gray')
    axes[0].set_title('Clean Image')
    axes[0].axis('off')

    for i, scale in enumerate(scales):
        noisy = add_rayleigh_noise(test_img, scale=scale)
        axes[i+1].imshow(noisy, cmap='gray')
        axes[i+1].set_title(f'Rayleigh Ïƒ={scale}')
        axes[i+1].axis('off')

    plt.tight_layout()
    plt.show()
```

```python
# Cell 5
class RayleighNoiseDataset(Dataset):
    def __init__(self, clean_images, noise_scales):
        self.clean_images = clean_images
        self.noise_scales = noise_scales
        # Total length = num_images * num_noise_scales
        self.total_length = len(clean_images) * len(noise_scales)

    def __len__(self):
        return self.total_length

    def __getitem__(self, idx):
        # Determine which image and which noise scale
        img_idx = idx // len(self.noise_scales)
        noise_idx = idx % len(self.noise_scales)

        # Get clean image
        clean_img = self.clean_images[img_idx]

        # Generate Rayleigh noisy image on-the-fly
        scale = self.noise_scales[noise_idx]
        noisy_img = add_rayleigh_noise(clean_img, scale=scale)

        # Convert to tensors
        clean_tensor = torch.FloatTensor(clean_img).unsqueeze(0)  # (1, H, W)
        noisy_tensor = torch.FloatTensor(noisy_img).unsqueeze(0)  # (1, H, W)

        return noisy_tensor, clean_tensor
```

```python
# Cell 6: Model Architecture
# --------------------------
class UltrasoundAutoencoder(nn.Module):
    def __init__(self):
        super(UltrasoundAutoencoder, self).__init__()
        # Encoder
        self.encoder = nn.Sequential(
            nn.Conv2d(1, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, 2),
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, 2),
        )
        # Decoder
        self.decoder = nn.Sequential(
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Upsample(scale_factor=2, mode='nearest'),
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Upsample(scale_factor=2, mode='nearest'),
            nn.Conv2d(64, 1, kernel_size=3, padding=1),
            nn.Sigmoid()
        )

    def forward(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded
```

```python
# Cell 7: Simplified Training Function (No Mixed Precision Issues)
# ---------------------------------------------------------------
def train_model_optimized(model, train_loader, criterion, optimizer, device, num_epochs=50):
    model.train()
    loss_history = []

    for epoch in range(num_epochs):
        epoch_loss = 0.0
        total_psnr = 0.0
        total_ssim = 0.0
        count = 0

        progress_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs}")

        for batch_idx, (noisy, clean) in enumerate(progress_bar):
            noisy, clean = noisy.to(device, non_blocking=True), clean.to(device, non_blocking=True)

            optimizer.zero_grad()

            # Standard forward pass (no mixed precision to avoid errors)
            outputs = model(noisy)
            loss = criterion(outputs, clean)

            # Standard backward pass
            loss.backward()
            optimizer.step()

            epoch_loss += loss.item()

            # Compute metrics less frequently for speed (every 50 batches)
            if batch_idx % 50 == 0 and batch_idx > 0:
                with torch.no_grad():
                    # Only compute for first sample in batch for maximum speed
                    output_sample = outputs[0, 0].cpu().numpy()
                    clean_sample = clean[0, 0].cpu().numpy()

                    psnr = peak_signal_noise_ratio(clean_sample, output_sample, data_range=1.0)
                    ssim = structural_similarity(clean_sample, output_sample, data_range=1.0)
                    total_psnr += psnr
                    total_ssim += ssim
                    count += 1

            # Update progress bar every 25 batches for better performance
            if batch_idx % 25 == 0:
                current_psnr = total_psnr / max(count, 1)
                current_ssim = total_ssim / max(count, 1)
                progress_bar.set_postfix({
                    'Loss': f'{loss.item():.4f}',
                    'PSNR': f'{current_psnr:.1f}',
                    'SSIM': f'{current_ssim:.3f}'
                })

        avg_loss = epoch_loss / len(train_loader)
        avg_psnr = total_psnr / max(count, 1)
        avg_ssim = total_ssim / max(count, 1)
        loss_history.append(avg_loss)

        print(f"Epoch [{epoch+1}/{num_epochs}] Loss: {avg_loss:.6f} | PSNR: {avg_psnr:.2f} dB | SSIM: {avg_ssim:.4f}")

        # Clear cache every 10 epochs
        if epoch % 10 == 0 and torch.cuda.is_available():
            torch.cuda.empty_cache()

    return loss_history
```

```python
# Cell 8: Rayleigh Noise Evaluation Function
# -------------------------------------------
def evaluate_rayleigh_denoising(model, test_images, device, scale, num_examples=3):
    """Evaluate model performance on Rayleigh noise at specific scale"""
    model.eval()
    psnr_list = []
    ssim_list = []

    idxs = random.sample(range(len(test_images)), min(num_examples, len(test_images)))

    fig, axes = plt.subplots(3, num_examples, figsize=(4 * num_examples, 8))
    if num_examples == 1:
        axes = axes.reshape(-1, 1)

    for i, idx in enumerate(idxs):
        clean = test_images[idx]
        noisy = add_rayleigh_noise(clean, scale=scale)

        noisy_tensor = torch.FloatTensor(noisy).unsqueeze(0).unsqueeze(0).to(device)

        with torch.no_grad():
            denoised_tensor = model(noisy_tensor)
            denoised = denoised_tensor.squeeze().cpu().numpy()

        psnr = peak_signal_noise_ratio(clean, denoised, data_range=1.0)
        ssim = structural_similarity(clean, denoised, data_range=1.0)
        psnr_list.append(psnr)
        ssim_list.append(ssim)

        axes[0, i].imshow(clean, cmap='gray')
        axes[0, i].set_title('Clean Image')
        axes[0, i].axis('off')

        axes[1, i].imshow(noisy, cmap='gray')
        axes[1, i].set_title(f'Rayleigh Noise (Ïƒ={scale})')
        axes[1, i].axis('off')

        axes[2, i].imshow(denoised, cmap='gray')
        axes[2, i].set_title(f'Denoised\nPSNR: {psnr:.2f} dB\nSSIM: {ssim:.3f}')
        axes[2, i].axis('off')

    plt.suptitle(f'Rayleigh Noise Denoising Results (Ïƒ={scale})')
    plt.tight_layout()
    plt.show()

    return np.mean(psnr_list), np.mean(ssim_list)
```

```python
# Data Loader Script for Summer_Project_25/Data.zip
# ===================================================

import os
import zipfile
from google.colab import drive
from tqdm import tqdm
import shutil

def load_ultrasound_data_from_drive():
    """
    Complete script to mount Google Drive, extract data.zip, and load ultrasound images
    """

    print("="*60)
    print("ULTRASOUND DATA LOADER")
    print("="*60)

    # Step 1: Mount Google Drive
    print("ðŸ”— Step 1: Mounting Google Drive...")
    if not os.path.exists('/content/drive'):
        drive.mount('/content/drive')
        print("âœ… Google Drive mounted successfully!")
    else:
        print("âœ… Google Drive already mounted")

    # Step 2: Locate the ZIP file
    print("\nðŸ“ Step 2: Locating data.zip file...")
    zip_path = "/content/drive/MyDrive/Summer_Project_25/Data/Data.zip"

    if not os.path.exists(zip_path):
        print(f"âŒ File not found: {zip_path}")

        # Check if Summer_Project_25 folder exists
        summer_project_path = "/content/drive/MyDrive/Summer_Project_25"
        if os.path.exists(summer_project_path):
            print(f"ðŸ“‚ Contents of Summer_Project_25 folder:")
            files = os.listdir(summer_project_path)
            for file in files:
                print(f"  ðŸ“„ {file}")

            # Look for any zip files
            zip_files = [f for f in files if f.lower().endswith('.zip')]
            if zip_files:
                print(f"\nðŸ” Found zip files: {zip_files}")
                # Use the first zip file found
                zip_path = os.path.join(summer_project_path, zip_files[0])
                print(f"âœ… Using: {zip_path}")
            else:
                print("âŒ No zip files found in Summer_Project_25 folder")
                return False
        else:
            print(f"âŒ Summer_Project_25 folder not found")
            print("Available folders in MyDrive:")
            mydrive_path = "/content/drive/MyDrive"
            if os.path.exists(mydrive_path):
                for item in os.listdir(mydrive_path):
                    item_path = os.path.join(mydrive_path, item)
                    if os.path.isdir(item_path):
                        print(f"  ðŸ“ {item}")
            return False
    else:
        print(f"âœ… Found zip file: {zip_path}")

    # Step 3: Check file size
    print("\nðŸ“Š Step 3: Checking file size...")
    file_size_gb = os.path.getsize(zip_path) / (1024**3)
    print(f"File size: {file_size_gb:.2f} GB")

    # Step 4: Check available space
    print("\nðŸ’¾ Step 4: Checking available disk space...")
    total, used, free = shutil.disk_usage("/content/")
    free_gb = free / (1024**3)
    print(f"Available space: {free_gb:.1f} GB")

    if free_gb < file_size_gb * 2:  # Need 2x space for zip + extracted
        print("âš ï¸  Warning: Low disk space. Extraction may fail.")
        print("Consider deleting unnecessary files or using a smaller dataset.")

    # Step 5: Extract the ZIP file
    print("\nðŸ“¦ Step 5: Extracting data.zip...")
    extraction_path = "/content/"

    # Remove existing Data folder if it exists
    data_folder = "/content/Data"
    if os.path.exists(data_folder):
        print("ðŸ—‘ï¸  Removing existing Data folder...")
        shutil.rmtree(data_folder)

    try:
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            file_list = zip_ref.namelist()
            print(f"Total files to extract: {len(file_list)}")

            # Extract with progress bar
            for file in tqdm(file_list, desc="Extracting", unit="files"):
                zip_ref.extract(file, extraction_path)

        print("âœ… Extraction completed successfully!")

    except zipfile.BadZipFile:
        print("âŒ Error: Invalid or corrupted zip file")
        return False
    except Exception as e:
        print(f"âŒ Error during extraction: {e}")
        return False

    # Step 6: Verify extracted data structure
    print("\nðŸ” Step 6: Verifying data structure...")

    if not os.path.exists(data_folder):
        print("âŒ Data folder not found after extraction")
        print("Available folders in /content/:")
        for item in os.listdir("/content/"):
            if os.path.isdir(f"/content/{item}"):
                print(f"  ðŸ“ {item}")
        return False

    # Check subdirectories
    subdirs = []
    for item in os.listdir(data_folder):
        item_path = os.path.join(data_folder, item)
        if os.path.isdir(item_path):
            subdirs.append(item)
            file_count = len(os.listdir(item_path))
            print(f"ðŸ“ Data/{item}: {file_count} files")

    if not subdirs:
        print("âš ï¸  No subdirectories found in Data folder")
        # List files directly in Data folder
        files = os.listdir(data_folder)
        print(f"Files directly in Data/: {len(files)}")
        if files:
            print(f"Sample files: {files[:5]}")

    print("\nâœ… Data structure verified!")
    return True

def load_ultrasound_images():
    """
    Load ultrasound images from the extracted data
    """
    print("\n" + "="*60)
    print("LOADING ULTRASOUND IMAGES")
    print("="*60)

    # Import required functions (make sure these are defined)
    from PIL import Image
    import numpy as np

    def load_ultrasound_data(data_path, image_size=(128, 128)):
        """Load ultrasound .tif images and exclude mask files"""

        if not os.path.exists(data_path):
            print(f"âŒ Path not found: {data_path}")
            return np.array([])

        # Get all .tif files in the directory
        all_files = [f for f in os.listdir(data_path) if f.lower().endswith('.tif')]

        # Filter out mask files (files containing '_mask' in the filename)
        scan_files = [f for f in all_files if '_mask' not in f.lower()]
        mask_files = [f for f in all_files if '_mask' in f.lower()]

        # Sort for consistent ordering
        scan_files = sorted(scan_files)

        print(f"Directory: {data_path}")
        print(f"Found {len(all_files)} total .tif files")
        print(f"  - Scan files: {len(scan_files)} (will be loaded)")
        print(f"  - Mask files: {len(mask_files)} (will be excluded)")

        if len(scan_files) == 0:
            print("âŒ No scan files found!")
            return np.array([])

        # Show sample filenames for verification
        if scan_files:
            print(f"Sample scan files: {scan_files[:5]}")

        images = []
        skipped_files = []

        print(f"Loading {len(scan_files)} scan images...")

        for img_name in tqdm(scan_files, desc="Loading images"):
            try:
                img_path = os.path.join(data_path, img_name)
                img = Image.open(img_path).convert('L')  # Convert to grayscale
                img = img.resize(image_size, Image.BILINEAR)
                arr = np.asarray(img, dtype=np.float32) / 255.0  # Normalize to [0,1]
                images.append(arr)
            except Exception as e:
                print(f"Error loading {img_name}: {e}")
                skipped_files.append(img_name)
                continue

        if skipped_files:
            print(f"âš ï¸  Skipped {len(skipped_files)} files due to errors")

        print(f"âœ… Successfully loaded {len(images)} ultrasound images")
        if images:
            print(f"Image shape: {images[0].shape}")

        return np.array(images)

    # Define paths
    train_path = "Data/train"
    test_path = "Data/test"

    # Load training images
    print("ðŸ“š Loading training images...")
    train_images = load_ultrasound_data(train_path, image_size=(128, 128))

    # Load test images
    print("\nðŸ§ª Loading test images...")
    if os.path.exists(test_path):
        test_images = load_ultrasound_data(test_path, image_size=(128, 128))
    else:
        print(f"âš ï¸  Test path not found: {test_path}")
        print("Using a subset of training images for testing")
        if len(train_images) > 100:
            test_images = train_images[-100:].copy()  # Last 100 for testing
            train_images = train_images[:-100]        # Rest for training
            print(f"Split: {len(train_images)} train, {len(test_images)} test")
        else:
            test_images = train_images.copy()
            print(f"Small dataset: using same {len(test_images)} images for both")

    # Store in global variables
    globals()['train_images'] = train_images
    globals()['test_images'] = test_images

    print("\n" + "="*60)
    print("LOADING SUMMARY")
    print("="*60)
    print(f"âœ… Training images: {len(train_images)}")
    print(f"âœ… Test images: {len(test_images)}")
    print(f"ðŸ“ Image shape: {train_images[0].shape if len(train_images) > 0 else 'No images'}")
    print(f"ðŸŽ¯ Images normalized to [0,1] range")
    print(f"ðŸš« Mask files automatically excluded")
    print("="*60)

    return train_images, test_images

# Main execution function
def main():
    """
    Main function to extract and load all data
    """
    print("ðŸš€ Starting ultrasound data loading process...")

    # Step 1: Extract data from Google Drive
    success = load_ultrasound_data_from_drive()

    if not success:
        print("âŒ Failed to extract data. Please check your file paths.")
        return None, None

    # Step 2: Load images into workspace
    train_images, test_images = load_ultrasound_images()

    if len(train_images) == 0:
        print("âŒ No training images loaded. Please check your data structure.")
        return None, None

    print("ðŸŽ‰ Data loading completed successfully!")
    print("âœ… Variables 'train_images' and 'test_images' are now available in your workspace")

    return train_images, test_images

# Execute the main function
if __name__ == "__main__":
    train_images, test_images = main()
```
